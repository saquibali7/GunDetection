```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
from IPython.display import Image
from random import choice
import shutil


data_path = 'input/labeled-guns-data-for-object-detection/data'
images = []
text = []

for (dirname, dirs, files) in os.walk(data_path):
    for file in files:
        if file.endswith('.txt'):
            text.append(file)
        else :
            images.append(file)
```

```python
!pwd
```

```python
print(len(text))
print(len(images))
```

```python
train_img = 'tempr/dataset/images/train'
test_img = 'tempr/dataset/images/valid'

train_txt = 'tempr/dataset/labels/train'
test_txt = 'tempr/dataset/labels/valid'

images_data_path = 'input/labeled-guns-data-for-object-detection/data/images'
text_data_path = 'input/labeled-guns-data-for-object-detection/data/labels'



for x in range(2000):
    img_file = choice(images)
    txt_file = img_file[:-4]+'.txt'
    
    shutil.copy(os.path.join(images_data_path, img_file), os.path.join(train_img, img_file))
    shutil.copy(os.path.join(text_data_path, txt_file), os.path.join(train_txt, txt_file))
    images.remove(img_file)
    text.remove(txt_file)
    

  
```

```python
test_img = 'tempr/dataset/images/valid'
test_txt = 'tempr/dataset/labels/valid'
for x in range(500):
    img_file = choice(images)
    txt_file = img_file[:-4]+'.txt'
    
    shutil.copy(os.path.join(images_data_path, img_file), os.path.join(test_img, img_file))
    shutil.copy(os.path.join(text_data_path, txt_file), os.path.join(test_txt, txt_file))
    images.remove(img_file)
    text.remove(txt_file)
```
